```python
from transformers import pipeline

# Load a text generation model that supports instruction-following
classifier = pipeline("text-generation", model="mistralai/Mistral-7B-Instruct-v0.1", max_new_tokens=10)
```

```python
def classify_email(text):
    prompt = (
        f"You are a helpful assistant who only answers with 'spam' or 'ham'.\n"
        f"Classify this email:\n\"\"\"\n{text.strip()}\n\"\"\"\nAnswer:"
    )
    response = classifier(prompt)[0]['generated_text']
    
    # Extract only the last token or class
    prediction = response.strip().split("Answer:")[-1].strip().lower()
    return prediction
```

```python
df["predicted_label"] = df["text"].apply(classify_email)
```
